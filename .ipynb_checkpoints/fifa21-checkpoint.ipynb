## changing datatype of joined to datetime

```python
player_details.Joined.head()
```

```python
def convert_month(x):
    if x == 'Jan':
        return 1
    elif x == 'Feb':
        return 2
    elif x == 'Mar':
        return 3
    elif x == 'Apr':
        return 4
    elif x == 'May':
        return 5
    elif x == 'Jun':
        return 6
    elif x == 'Jul':
        return 7
    elif x == 'Aug':
        return 8
    elif x == 'Sep':
        return 9
    elif x == 'Oct':
        return 10
    elif x == 'Nov':
        return 11
    elif x == 'Dec':
        return 12
```



  ***function to change datetime***

```python
date = []
for x in range(len(player_details['Joined'])):
    d = player_details['Joined'][x]
    c = d.split(" ")
    month = str(convert_month(c[0]))
    day = str(c[1].replace(',',''))
    year = str(c[2])
    if len(day) == 1:
        day = ('0'+str(day))
    date_long = (str(month)+'/'+str(day)+'/'+str(year))
    date_con = pd.to_datetime(date_long)
    date.append(date_con)
player_details['Joined'] = date
player_details['Joined']
```

## change datatype of shots and remove k values and multiply it with 1000

```python
def shot_on_target(x):
    if 'K' in x:
        x = x.replace('K','') 
        x = float(x)*1000
        return x
    else:
        return x
player_abilities['Shots'] = player_abilities['Shots'].apply(shot_on_target)
    
        
```

### removing \n from club

```python
def invalid_format(x):
    if '\n\n\n\n' in x:
        y = x.replace('\n\n\n\n','')
        return y
    else:
        return x



player_details['Club'] = player_details['Club'].apply(invalid_format)
```

```python
player_details['Club'].tail(30)
```

### change contract column status to ACTIVE, LOAN, FREE

```python
player_details.Contract.unique()
```

```python
def contract_length(a):
    if "On Loan" in a:
        a = 'ON LOAN'
        return a
    elif '~' in a:
        a = 'ACTIVE'
        return a
    elif 'Free' in a:
        a = "FREE"
        return a


player_details['Contract'] = player_details['Contract'].apply(contract_length)
```

```python
player_details.Contract.unique()
```

```python
player_details.head()
```

#### change wage into int and remove k and euro sign

```python
player_details.Wage.unique()
```

```python
def wage_change(x):
    if 'K' in x:
        x = x.replace('K','')
        x = x.replace('€','')
        x = int(x)*1000
        return x
    else:
        x = x.replace('€','')
            
        return x
        

        
player_details['Wage'] = player_details['Wage'].apply(wage_change)       
```

```python
player_details.Wage.unique()
```

```python
player_details['Wage'] = player_details['Wage'].astype('int64')
```

```python
player_details = player_details.rename(columns={'Wage':'Wage €'})
```

```python
player_details.head()
```

```python
player_details.Value.unique()
```

```python
def Value_change(x):
    if 'M' in x:
        x = x.replace('M','')
        x = x.replace('€','')
        x = float(x)*100000
        return x
    elif 'K' in x:
        x = x.replace('K','')
        x = x.replace('€','')
        x = float(x)*1000
        return x
    else:
        x = x.replace('€','')
            
        return x
        

        
player_details['Value'] = player_details['Value'].apply(Value_change)
```
